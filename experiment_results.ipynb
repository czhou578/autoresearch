```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the results
df = pd.read_csv('results.tsv', sep='\t')

# Plot validation loss vs number of experiments
plt.figure(figsize=(10, 6))
plt.plot(range(len(df)), df['loss'], marker='o', linestyle='-', color='b')

plt.title('Validation Loss vs. Number of Experiments (mar14)')
plt.xlabel('Number of Experiments')
plt.ylabel('Validation Loss')
plt.grid(True)

for i, row in df.iterrows():
    plt.annotate(row['description'], (i, row['loss']), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
