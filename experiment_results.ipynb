```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('results.tsv', sep='\t')

# Filter crashes (where loss is 0.0) if any
df = df[df['loss'] > 0.0]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(df) + 1), df['loss'], marker='o', linestyle='-', color='b')
plt.title('Validation Loss vs Number of Experiments')
plt.xlabel('Number of Experiments Taken')
plt.ylabel('Validation Loss')
plt.grid(True)
plt.show()
```
